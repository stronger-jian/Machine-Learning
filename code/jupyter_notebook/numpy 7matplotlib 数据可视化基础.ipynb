## matplotlib 数据可视化基础

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
cosy=np.cos(x)
```

```python
siny=y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
```

```python
x1=np.linspace(0,10,10)
cosy=np.cos(x1)
siny=np.sin(x1)
plt.plot(x1,siny)
plt.plot(x1,cosy)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle=":")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-.")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-")
plt.xlim(-5,15)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-")
plt.ylim((-0, 1))
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-")
plt.axis([-1,11,-2,2])
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="-")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="-")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="-")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("welcome to study matplotlib")
```
